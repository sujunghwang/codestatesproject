```python
import os
from glob import glob
import xml.etree.ElementTree as et
import re
from pykospacing import Spacing

import kss
```

```python
li = glob('drive/MyDrive/cp2/신문원문PDF/*.xml')
li.sort()
li
```

```python
text = []
for i in li:    
  try:
    doc = et.parse(i)

    root = doc.getroot()

    t = root.find("NewsItem").find("NewsComponent").find("NewsComponent").find("ContentItem").find("DataContent").text
    t = re.sub('[\n\t]', '', t)
    text.append(t)
  except:
    print(i)
```

```python
text.remove('전면광고')
text.remove('지면광고')
text = list(text)
text
```

```python
del text[122]
del text[96]
del text[75]
del text[10]

text
```

```python
spacing = Spacing()
kospacing_sent = spacing(text[2]) 

print(text[2])
print(kospacing_sent)
```

```python
sentences = []
s_sentences = []

for sentence in text:
    
    sentences.append(sentence)

    sentence = sentence.replace("'", "")
    sentence = sentence.replace("■", "")
    sentence = sentence.replace("●", "")
    sentence = sentence.replace("◇", "")
    sentence = sentence.replace("△", "")
    sentence = sentence.replace(" ", "")

    s_sentences.append(sentence)  

s_sentences
```

```python
for i in range(len(s_sentences)):
  kospacing_sent = spacing(s_sentences[i]) 

  print("원문:")
  print(sentences[i])
  print("PyKoSpacing: ")
  print(kospacing_sent)
  print()
```
